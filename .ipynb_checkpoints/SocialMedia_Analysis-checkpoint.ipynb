## Task 3: Data retrieval and cleaning

* For this part of the process, we used the API key and the appopriate url to retrieve Revenues and Stock prices for **Facebook Inc, Twitter, Sina, Snap Inc and Pinterest**
* We are retriving data for the last 5 quarters and thus, the period parameter for the url is set to 5 to give the right information
* The information is returned is a json format and is stored in dictionaries created for each company
* The data is later appended to a main dictionary that is going to be used for the rest of the process


```python
#Url from financial modeling prep to grab quarterly financial statement
base_url = "https://financialmodelingprep.com/api/v3/income-statement/"
    
#List of parameters for request 
companies =["SNAP", "TWTR","FB", "SINA","PINS"]
limit = 5
cmny_info = []
time = "quarter"

#Mian dictionary and sub dictionnary to store json response
main_dict = {
    "SNAP":{
    "Dates" :[],
    "Revenue":[],
    'Stock Price':[]}, 

    "TWTR":{
    "Dates" :[],
    "Revenue":[],
    'Stock Price':[]},

    "FB":{
    "Dates" :[],
    "Revenue":[],
    'Stock Price':[]},

    "SINA":{
    "Dates" :[],
    "Revenue":[],
    'Stock Price':[]},

    "PINS":{
    "Dates" :[],
    "Revenue":[],
    'Stock Price':[]},

}
# loop through all the companies and append to at the end
for company in companies:
    
    #Create search query, make request and store in json
    url = f"https://financialmodelingprep.com/api/v3/income-statement/{company}\
?period=quarter&limit=5&apikey={api_key}"
    
    #Try and except method to get all data
    try:
        response = requests.get(url).json()
    except:
        print ("Sorry, we don't have this ticker: " + company)
    for i in range(0,5):
            main_dict[company]['Dates'].append(response[i]['date'])
            main_dict[company]['Revenue'].append(response[i]['revenue'])
            
#print main dictitionary to verify data
pprint (main_dict)
```

## Task 4: Using the Pandas Library and yfinance for additional Data Modification

* yfinance was used to retrieve **stock prices (Open price)** for each company using the ticker
* Companies list was outlined under Task 3
* The period was set to 2 years to get historical stock information from the yahoo finance depency
* The information received displays every single date for the last 2 years
* The data was was cleaned by dropping the indexes to keep specific data on the open price
* Since this analysis is based on quarters, the dates from the main dictionary aboved were used as a reference to filter through each row and find the open price matching that date
* To add stock prices, the open prices was appended to the main dictionary   
* There are now 4 columns for the analysis, **Date, Symbol,Revenue and stock price**

```python
#Load stock price information from yfinance
tickerdf = yf.download(tickers = companies,period = '2y',group_by = 'ticker',auto_adjust = True)

#Drop index to keep the data on the open prices(stock prices)
snap = tickerdf['SNAP'].reset_index(drop=False)
fbk = tickerdf['FB'].reset_index(drop=False)
twt = tickerdf['TWTR'].reset_index(drop=False)
sina = tickerdf['SINA'].reset_index(drop=False)
pins = tickerdf['PINS'].reset_index(drop=False)

#Use the quarters dates from the main dictionary to match the dates of the open prices
dates = ['9/30/2019','12/31/2019','3/31/2020','6/30/2020','9/30/2020']

#Filter throught the rows to get the date that matches the quarters dates
snap = snap.loc[snap['Date'].isin(dates)]
fbk = fbk.loc[fbk['Date'].isin(dates)]
twt = twt.loc[twt['Date'].isin(dates)]
sina = sina.loc[sina['Date'].isin(dates)]
pins = pins.loc[pins['Date'].isin(dates)]

#add stock and df to get the open price from yfinance and append to the main dictionary
def add_stock(df,company):
    for item in list(df['Open']):
        main_dict[company]['Stock Price'].append(item)
        
#Add all data together 
add_stock(snap,'SNAP')
add_stock(fbk,'FB')
add_stock(twt,'TWTR')
add_stock(sina,'SINA')
add_stock(pins,'PINS')

#Print one variable to see how open price(Stock prices) is being displayed
print(snap)
```

## II: Revenues Analysis 

* Revenues analysis from the main dictionary
* The first part consisted of isolating the revenues for each ticker symbol
* The dates were sorted from the oldest to the latest for better plotting
* Pandas matplotlib was used to plot the data
* Diffrent colours were picked for each label
* The graph below is a visual representation of the revenues for each ticker with their quaterly dates
* Some revenues are in billions, to better represent the amounts the revenues were divided by 1 million

```python
#Locate revenue data for all companies and store in variables
sp1 = main_dict["SNAP"]['Revenue']
tw1 = main_dict["TWTR"]['Revenue']
fb1 = main_dict["FB"]['Revenue']
sina1 = main_dict["SINA"]['Revenue']
pint1= main_dict["PINS"]['Revenue']
```

```python
#Divide the revenues by 1 million to ease plotting in $M
myInt = 1000000
snaplist = [x / myInt for x in sp1]
twitlist = [x / myInt for x in tw1]
fblist = [x / myInt for x in fb1]
silist = [x / myInt for x in sina1]
pilist = [x / myInt for x in pint1]
```

```python
#Plot the data using pandas library
dates = ['9/30/2019','12/31/2019','3/31/2020','6/30/2020','9/30/2020']
w = 0.1
bar1 = np.arange(len(dates))
bar2 = [i + w for i in bar1]
bar3 = [i + w for i in bar2]
bar4 = [i + w for i in bar3]
bar5 = [i + w for i in bar4]

fig1,ax1 = plt.subplots(figsize=(18, 10))
snapb = plt.bar(bar1,snaplist,w, color ='red', label ="SNAP")
twib= plt.bar(bar2,twitlist,w, color ='blue', label = "TWTR")
fbb = plt.bar(bar3,fblist,w, color ='green', label = "FB")
sib = plt.bar(bar4,silist,w, color ='purple', label = "SINA")
pinb = plt.bar(bar5,pilist,w, color ='darkorange', label = "PINS")
plt.xticks(bar1+w/2,dates)
plt.ylim(0,25000)
plt.grid()
plt.xlabel('Dates')
plt.ylabel(f'Revenues ($B/M)')
plt.title(f' Revenues Over Quarters')
plt.legend(handles=[snapb, twib,fbb,sib,pinb], loc="best")


def insert_data_labels(bars):
	for bar in bars:
		bar_height = bar.get_height()
		plt.annotate('{0:.0f}'.format(bar.get_height()),
			xy=(bar.get_x() + bar.get_width() / 2, bar_height),
			xytext=(0, 3),
			textcoords='offset points',
			ha='center',
			va='bottom'
		)

insert_data_labels(snapb)
insert_data_labels(twib)
insert_data_labels(fbb)
insert_data_labels(sib)
insert_data_labels(pinb)

plt.savefig("./output_data/figure_3.png")
```
